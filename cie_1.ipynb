```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/mtcars (2).csv')
print(df)
df.head()
df.tail()
df.describe()
```

```python
import matplotlib.pyplot as plt
auto_count=(df['am']==0).sum()
manual_count=(df['am']==1).sum()
labels=['Automatic','manual']
counts=[auto_count,manual_count]
plt.bar(labels,counts,color=['orange','blue'])
plt.title("frequency of transmission")
plt.xlabel("transmission")
plt.ylabel("numbers of cars")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=df['wt']
y=df['mpg']
plt.scatter(x,y)
plt.title("car weight vs mpg")
plt.xlabel("weight")
plt.ylabel("mpg")
plt.show()
```

```python
import matplotlib.pyplot as plt
y=df['mpg']
plt.boxplot(y)
plt.title("boxplot of mpg")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=df['mpg']
bins=8
plt.hist(x)
plt.title("histogram of mpg")
plt.xlabel("mpg")
plt.ylabel("frequency")
plt.show()
```
